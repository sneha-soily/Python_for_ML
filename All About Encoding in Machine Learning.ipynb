```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#Handle Null value

mean = df.Transport.mean()
mean
```

```python
median = df.Transport.median()
median
```

```python
#use any value like mean or median

df.Transport = df.Transport.fillna(median) #if nan value, then fill  
```

```python
df.isnull().sum()
```

# Encoding

```python
#manually replace

df.Area = df.Area.replace(['Dhaka', 'Ctg', 'Rangpur'], [3,2,1])
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
label = LabelEncoder()
```

```python
df.Area = label.fit_transform(df['Area'])
```

```python
df.head()
```

```python
#we are use in this way by loop

for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
df.Area.head()
```

# One_Hot Encoding

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
pd.get_dummies(df['Area'])
```

```python
#drop colomn ctg
dummy_variables = pd.get_dummies(df['Area'],drop_first=True)
```

```python
dummy_variables.head()
```

```python
df.head()
```

```python
new_df = df.drop('Area',axis=1)
```

```python
new_df.head()
```

```python
#new dataset and old data set concatenet

df = pd.concat([new_df,dummy_variables], axis=1)
```

# Ordinal Encoder

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.Area.unique()
```

```python
city_list = ['Dhaka', 'Ctg', 'Rangpur']
```

```python
ordinal = OrdinalEncoder(categories=[city_list])
```

```python
encode_values = ordinal.fit_transform(df[['Area']]) #number of sample & number of feature
```

```python
encode_values
```

```python
new_area = pd.DataFrame(encode_values, columns=['Area'])
```

```python
new_area
```

```python
 df.Area = new_area
```

```python
df.head()
```
